```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
df = pd.read_csv(r"C:\Users\itsbe\NLP_Bets\NLP_Bets\coletas\resultados\analise_nlp.csv")

def contar_palavras(coluna):
    todas = ", ".join(df[coluna].dropna()).split(", ")
    return Counter(todas)

# Gráfico 1
substantivos_freq = contar_palavras("Substantivos").most_common(10)
sub_palavras, sub_freq = zip(*substantivos_freq)

plt.figure(figsize=(10, 5))
plt.barh(sub_palavras[::-1], sub_freq[::-1], color="steelblue")
plt.title("Top 10 Substantivos Mais Frequentes")
plt.xlabel("Frequência")
plt.tight_layout()
plt.show() 
```

```python
# Gráfico 2
entidades_freq = contar_palavras("Entidades").most_common(10)
ent_textos, ent_freq = zip(*entidades_freq)

plt.figure(figsize=(10, 5))
plt.barh(ent_textos[::-1], ent_freq[::-1], color="orange")
plt.title("Top 10 Entidades Nomeadas")
plt.xlabel("Frequência")
plt.tight_layout()
plt.show()
```
